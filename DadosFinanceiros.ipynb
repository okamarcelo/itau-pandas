```python
import sqliteDb as db
from itables import init_notebook_mode, show
import pandas as pd
import datetimeFunc as dt
from datetime import datetime
import numpy as np
import itauExcel as i

init_notebook_mode(all_interactive=False)

def TransformarDataFrame(dataFrame):
    return pd.DataFrame({'Data': [datetime.fromtimestamp(d) for d in dataFrame.TransactionEpoch], 
                  'Lançamento': dataFrame.TransactionTitle, 
                  'Categoria': dataFrame.TransactionCategory, 
                  'Origem': dataFrame.TransactionOrigin,
                  'Período': [datetime.strptime(d, "%Y%m") for d in dataFrame.TransactionPeriod],
                  'Valor': dataFrame.TransactionAmount})

def ImportarExcel(periodo):
    return i.ImportExcel('~/Downloads/' + periodo + '.xls', periodo, 'Credito')

def ImportarDatabase(periodo):
    return db.SelectData("finantialData.sqlite", 
                         "Select * from Transactions where TransactionPeriod = " + periodo)

def GerarMascaraHash(df):
    return pd.util.hash_pandas_object(pd.DataFrame({'TransactionAmount':df.TransactionAmount.astype('float64'), 
                                                    'TransactionEpoch':df.TransactionEpoch.astype('int64'), 
                                                    'TransactionTitle':df.TransactionTitle}),index=False)

def EncontrarDiferencas(d1,d2):
    res = d1.where(~(GerarMascaraHash(d2).isin(GerarMascaraHash(d1))))
    return res[res.TransactionTitle.notnull()] 

def Diff(d1,d2):
    return pd.concat([EncontrarDiferencas(d1,d2),EncontrarDiferencas(d2,d1)])

periodos = ["202211", "202212"]

dxl = ImportarExcel(periodos[0])
ddb = ImportarDatabase(periodos[0])

d = Diff(dxl,ddb)

ddd = db.InsertIfNotExist('finantialData.sqlite', dxl, 'Transactions', 'UniqueColumns')

# db.SaveData('finantialData.sqlite', d, 'Transactions')


#dt = TransformarDataFrame(dd).pivot_table(values='Valor', index='Categoria', aggfunc=sum, columns='Período', margins=True)

#show(pt)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho completo do arquivo
caminho_arquivo = '~/Downloads/Planilhas/FinancasPessoais2023.xlsx'

# Carregando todas as planilhas em um dicionário de dataframes
dfs = pd.read_excel(caminho_arquivo, sheet_name=None)

# Criando um dataframe vazio para armazenar os gastos por categoria em cada mês
gastos_por_categoria = pd.DataFrame(columns=['Categoria'] + list(dfs.keys()))

# Preenchendo o dataframe com os gastos por categoria em cada mês
for sheet_name, df in dfs.items():
    # Extraindo o nome do mês da chave do dicionário
    mes_ano = sheet_name.strip()
    # Renomeando a coluna de gastos para o nome do mês
    gastos = df.groupby('Categoria')['Valor'].sum().reset_index()
    gastos.rename(columns={'Valor': mes_ano}, inplace=True)
    # Adicionando a coluna de gastos por categoria em cada mês ao dataframe principal
    gastos_por_categoria = pd.merge(gastos_por_categoria, gastos, on='Categoria', how='outer')

# Calculando a média dos gastos por categoria em todos os meses
media_gastos_por_categoria = gastos_por_categoria.copy()

```
